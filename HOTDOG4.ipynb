```python
import numpy as np
```

```python
data = []
```

```python
categories = ['hotdog','nothotdog']
```

```python
# it will go to directory where file hotdog and nothotdog is stored and then it will take imageS from 
# hotdog and nothotdog and store it in 2D list named data, in this list each element will 
# have 2 elements in it first element will be image, second will be 0 or 1 (which denotes hotdog or nothotdog) 
# and it will also resize images into size (224,224)

for category in categories:
    path = os.path.join(r'C:\Users\CDMG-1\Desktop\Ujas Adepal\hotdog dataset\archive\hotdog-nothotdog\hotdog-nothotdog\train',category)
    label = categories.index(category)

    for file in os.listdir(path):
        img_path = os.path.join(path,file)
        img = cv2.imread(img_path)
        img = cv2.resize(img,(224,224))
        data.append([img,label])
```

```python
# it will separate image data and 0/1 (hotdog/nothotdog)

x = []
y = []
for features,label in data:
    x.append(features)
    y.append(label)
```

```python
# it will make list x and y into numpy array

x = np.array(x)
y = np.array(y)
```

```python
# it will bring image data into range of 0 to 1

x  = x/255
```

```python
# it will split train and test data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from keras.applications.vgg16 import VGG16
```

```python
# we have made VGG16 model which has 21 layers 
# thirteen convolutional layers, five Max Pooling layers, and three Dense layers

vgg = VGG16()
```

```python
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential()
```

```python
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.summary()
```

```python
from keras.layers import Dense
```

```python
# adding and extra dense layer on VGG16 for binary classification purpose

model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
# compiling model with optimizer as adam loss function as binary crossentropy and metrics as accuracy

model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# it will train model on x_train and y_train for 5 epochs with validation data as x_test and y_test
# with 5 epoches this model gave us accuracy of 82.20% 
# it took 28 minutes to train it on cpu 

model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
# it will save the weights of the model

model.save_weights(r'C:\Users\CDMG-1\Desktop\Ujas Adepal\hotdog dataset\archive\hotdog.h5')
```

```python
# this function will take image as an input and it will 
# return pridiction of image whether image is hotdog or not

def detect_hotdog(img):

    y_pred = (model.predict(img.reshape(1,224,224,3)) > 0.5).astype("int32")

    return y_pred[0][0]
```

```python
# used cv2 to capture video

cap= cv2.VideoCapture(0)
```

```python
# this will turn on camera and it will predict whether the thing infront of camera is hotdog or not
# if the thig is hotdog then it will print zero else 1

while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame,(224,224))
    y_pred = detect_hotdog(frame)
    print(y_pred)
    cv2.imshow("window",frame)
    if cv2.waitKey(1)& 0xFF == ord('x'):
        break
        
cv2.destroyAllWindows()
```
